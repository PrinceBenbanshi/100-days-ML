### Feature Scaling - Normalization 

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
#Theare are not column names in the dataset so we will add them manually
df=pd.read_csv('wine_data.csv',header=None,usecols=[0,1,2])
df.columns=['Class label','Alcohol','Malic acid']
df.head()
```

```python
sns.kdeplot(x=df['Alcohol'])
plt.show()
```

```python
sns.kdeplot(x=df['Malic acid'])
plt.show()
```

```python
colors_dict={1:'r',2:'g',3:'b'}
sns.scatterplot(x=df['Alcohol'],y=df['Malic acid'],hue=df['Class label'],palette=colors_dict)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df.drop('Class label',axis=1),df['Class label'],test_size=0.3,random_state=0)
x_train.shape,x_test.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
#fit the scaler only on the training data
scaler.fit(x_train)
#now transform both the training and testing data
x_train_scaled=scaler.transform(x_train)
x_test_scaled=scaler.transform(x_test)
```

```python
x_train_scaled=pd.DataFrame(x_train_scaled,columns=x_train.columns)
x_test_scaled=pd.DataFrame(x_test_scaled,columns=x_test.columns)
```

```python
np.round(x_train.describe(),2)
```

```python
np.round(x_train_scaled.describe(),2)
```

```python
fig,(ax1,ax2)=plt.subplots(ncols=2,figsize=(10,5))
ax1.scatter(x_train['Alcohol'],x_train['Malic acid'],c=y_train)
ax1.set_title('Before Scaling')
ax2.scatter(x_train_scaled['Alcohol'],x_train_scaled['Malic acid'],c=y_train)
ax2.set_title('After Scaling')
plt.show()
```

```python
# Visualizing the distribution of features before scaling
fig,(ax1,ax2)=plt.subplots(ncols=2,figsize=(10,5))
sns.kdeplot(x=x_train['Alcohol'],ax=ax1)
ax1.set_title('before scaling')
sns.kdeplot(x=x_train['Malic acid'],ax=ax1)
ax1.legend(['Alcohol','Malic acid'])
#Visualizing the distribution of features after scaling
sns.kdeplot(x=x_train_scaled['Alcohol'],ax=ax2)
ax2.set_title('after scaling')
sns.kdeplot(x=x_train_scaled['Malic acid'],ax=ax2)
ax2.legend(['Alcohol','Malic acid'])
plt.show()

```
